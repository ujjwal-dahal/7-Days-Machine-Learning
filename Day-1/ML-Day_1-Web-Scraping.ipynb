# **Let's scrap data from wikipedia**

```python
from bs4 import BeautifulSoup # for parsing HTML data
import requests # send HTTP requests
```

```python
url = 'https://en.wikipedia.org/wiki/Nepal'
```

```python
page = requests.get(url) #Esle Website ko HTML aaucha

soup = BeautifulSoup(page.text, 'html')
```

```python
if page.status_code == 200:
    print("Page fetched successfully!")
else:
    print(f"Failed to fetch page: {page.status_code}")
```

```python
# Display the soup value
print(soup); #HTML Data Display Huncha
```

```python
# Display any text of HTML
# use find and findall

#soup.find("h1")  #HTML bata Sab bhanda Mathiko h1 tag bhako dincha


#Sabai Find Garna
#soup.find_all("p") #Tyo Website ma use bhako sabai Paragraph Tag bhako dincha


```

```python
# display 'p' tag using find_all and use strip to remove unwanted space


all_para = soup.find_all("p")
for each_para in all_para:
  print(each_para.text.strip())
```

```python
# target all table tag and target our desired table (using class attribute)

table_data = soup.find("table",class_="wikitable sortable")
# table_data
```

```python
# fetch all the header of that tabel

header_table = table_data.find_all("th")
for each_header in header_table:
  print(each_header.text.strip())
```

```python
# convert that header into list
header_data = ["h1", "h2", "h3"]
content_data = [["c1","c2","c3"],["c4","c5","c6"]]
```

```python
# Remove unwanted header from that list

import pandas as pd

df = pd.DataFrame(content_data,columns=header_data) #columns le list matra accept garcha

# df
```

```python
# convert list into dataframe
df.to_csv("machinelearning.csv") #csv file ko name ma save garcha
```
